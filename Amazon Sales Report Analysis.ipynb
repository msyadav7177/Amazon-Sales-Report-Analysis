```python
Revenue_Report_for_category = data.groupby(['Category','Size'])['Amount'].sum()
Revenue_Report_for_category
```

```python
Revenue_Report_for_category = Revenue_Report_for_category.reset_index()
Revenue_Report_for_category
```

```python
plt.figure(figsize=(25,6))
Revenue_Report_for_category.pivot(index='Category', columns='Size', values='Amount').plot(kind='bar',stacked=True)
plt.title('Revenue Report by Category and Size')
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.legend(title='Size')
plt.show()
```

```python
data
```

```python
Revenue_report_by_Fulfilment = data.groupby('Fulfilment')['Amount'].sum()/100000
Revenue_report_by_Fulfilment
```

```python
plt.figure(figsize=(16,6))
sns.barplot(data=Revenue_report_by_Fulfilment,palette='viridis')
plt.title("Revenue report for a particular Fulfilment method")
plt.xlabel("Fulfilment method")
plt.ylabel("Total Revenue(in lac)")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_amount_order = data.groupby('Fulfilment')['Amount'].mean().sort_values(ascending=False)
avg_amount_order
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_amount_order.index, y=avg_amount_order.values, palette='Blues_r')
plt.title('Average Amount per Order for a particular Fulfillment Method')
plt.xlabel('Fulfillment Method')
plt.ylabel('Average Amount (in Rs.)')
plt.show()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['month'] = data['Date'].dt.month
data['month'].unique()
```

```python
months = ['march','April', 'May', 'June']
data['month'].replace([3,4,5,6],months,inplace=True)
```

```python
print(f'The earliest date is {data["Date"].min()}')
print(f'The latest date is {data["Date"].max()}')
```

```python
monthly_sales = data.groupby('month')['Amount'].sum()/100000
monthly_sales
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values, palette='viridis')
plt.title('Monthly Sales Distribution')
plt.xlabel('Month')
plt.ylabel('Total Sales ( in Lac )')
plt.show()
```
